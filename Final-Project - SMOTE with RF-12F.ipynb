```python
#Data Visualization
C= CC.columns.values
i = 0
C0 = CC.loc[CC['Class'] == 0]
C1 = CC.loc[CC['Class'] == 1]

sns.set_style('whitegrid')
plt.figure()
fig, ax = plt.subplots(8,4,figsize=(16,28))

for feature in C:
    i += 1
    plt.subplot(8,4,i)
    sns.kdeplot(C0[feature], bw=0.5,label="Class = 0")
    sns.kdeplot(C1[feature], bw=0.5,label="Class = 1")
    plt.xlabel(feature, fontsize=9)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=9)
plt.show()
```

```python
#Features scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
CC["S_Amount"]=  sc.fit_transform(CC.iloc[:,29].values.reshape(-1,1))
```

```python
CC.columns
```

```python
#Dependent and independent variables
Features_CC= CC[['V1', 'V2', 'V3', 'V4', 'V9', 'V10','V11', 'V12','V14','V18', 'V19', 'S_Amount']]
Labels_CC= CC[['Class']]
```

```python
#Train and test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Features_CC, Labels_CC, test_size = 0.25, random_state = 0)
```

```python
#Train and test split
print (x_train.shape)
print (y_train.shape)
print (x_test.shape)
print (y_test.shape)
```

```python
#Oversampling SMOTE
from imblearn.over_sampling import SMOTE
OS=SMOTE(random_state=0)
OS_x_train,OS_y_train=OS.fit_sample(x_train,y_train)

print (OS_y_train["Class"].value_counts())
print (OS_x_train.shape)
```

```python
#Balanced class visualization
plt.title('OS_y_train "Class" Plot')
sns.countplot(x='Class',data=OS_y_train)
```

```python
#Array formation
Nx=np.asarray(OS_x_train)
Ny=np.asarray(OS_y_train["Class"])
```

```python
#Training RF model
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier().fit(Nx,Ny) 
```
