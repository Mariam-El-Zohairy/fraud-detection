```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.title('CC "Class" Plot')
sns.countplot(x='Class',data=CC)
```

```python
sns.lmplot(x='Time',y='Amount',data=CC,fit_reg=False,hue='Class',height=7,aspect=2)
plt.title('CC Amount-Time Plot')
```

```python
plt.figure(figsize = (10,10))
plt.title('CC Correlation Plot')
sns.heatmap(CC.corr(),linewidths=0.02)
```

```python
C= CC.columns.values
i = 0
C0 = CC.loc[CC['Class'] == 0]
C1 = CC.loc[CC['Class'] == 1]

sns.set_style('whitegrid')
plt.figure()
fig, ax = plt.subplots(8,4,figsize=(16,28))

for feature in C:
    i += 1
    plt.subplot(8,4,i)
    sns.kdeplot(C0[feature], bw=0.5,label="Class = 0")
    sns.kdeplot(C1[feature], bw=0.5,label="Class = 1")
    plt.xlabel(feature, fontsize=9)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=9)
plt.show()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
CC["S_Amount"]=  sc.fit_transform(CC.iloc[:,29].values.reshape(-1,1))
```

```python
NF=CC[CC["Class"]==0][0:492]
F=CC[CC["Class"]==1][0:492]
Remaining_NF=CC[CC["Class"]==0][492:]
```

```python
Frames=[NF,F]
BCC=pd.concat(Frames)
BCC
```

```python
plt.title('BCC "Class" Plot')
sns.countplot(x='Class',data=BCC)
```

```python
sns.lmplot(x='Time',y='Amount',data=BCC,fit_reg=False,hue='Class',height=7,aspect=2)
plt.title('BCC Amount-Time Plot')
```

```python
plt.figure(figsize = (10,10))
plt.title('BCC Correlation Plot')
sns.heatmap(CC.corr(),linewidths=0.02)
```

```python
CB= BCC.columns.values
i = 0
C0 = BCC.loc[BCC['Class'] == 0]
C1 = BCC.loc[BCC['Class'] == 1]

sns.set_style('whitegrid')
plt.figure()
fig, ax = plt.subplots(8,4,figsize=(20,30))

for feature in CB:
    i += 1
    plt.subplot(8,4,i)
    sns.kdeplot(C0[feature], bw=0.5,label="Class = 0")
    sns.kdeplot(C1[feature], bw=0.5,label="Class = 1")
    plt.xlabel(feature, fontsize=9)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=9)
plt.show()
```

```python
F_1_BCC= BCC[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'S_Amount']]
x=np.asarray(F_1_BCC)
y=np.asarray(BCC['Class']) 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
print (x_train.shape)
print (y_train.shape)
print (x_test.shape)
print (y_test.shape)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

```

```python
from sklearn.metrics import confusion_matrix,f1_score
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
f1_score(y_test,y_pred)
```
